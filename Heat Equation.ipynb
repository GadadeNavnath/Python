# Solution of Heat Equation

## $u_{t}=u_{xx}$ with BCs $u(0,t)=u(1,t)=0 $

## IC is  $u(x,0)=\sin(\pi x);0<x<1 $

```python
import numpy as np
import matplotlib.pyplot as plt
h=0.01
k=0.025
x=np.arange(0,1+h,h).round(3)
t=np.arange(0,0.1+k,k).round(3)
BCs=[0,0]
IC=np.sin(np.pi*x)
m=len(t)
n=len(x)
T=np.zeros((n,m))
T[0,:]=BCs[0]
T[-1,:]=BCs[1]
T[:,0]=IC
factor=k/h**2
for j in range(1,m):
    for i in range(1,n-1):
        T[i,j]=factor*T[i-1,j-1]+(1-2*factor)*T[i,j-1]+factor*T[i+1,j-1]
T.round(3)
R=np.linspace(1,0,m)
B=np.linspace(0,1,m)
G=0
for j in range(m):
    plt.plot(x,T[:,j],color=[R[j],G,B[j]])
    plt.plot(x,np.exp(-(np.pi**2)*0.1)*np.sin(np.pi*x),"k-")
plt.legend([f't={value} s' for value in t],loc="upper right")
plt.xlabel("Distances(m)")
plt.ylabel("Temprature in [$\degree$ C]")
plt.xlim([0,1])
plt.ylim([0,1])
plt.grid()
```
